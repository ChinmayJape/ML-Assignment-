## Solution of Data Cleaning Assignment in Pandas

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN',
'londON_StockhOlm',
'Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']})
```

```python
df.head()
```

### Answer 1

```python
df.fillna(method='ffill')
```

```python
df = df.interpolate()
df
```

```python
convert_dict = {'FlightNumber': int} 
df = df.astype(convert_dict) 
print(df.dtypes)
```

### Answer 2

```python
df[['From','To']] = df.From_To.str.split("_",expand=True,)
df
```

### Answer 3

```python
df['From'] = df['From'].str.capitalize()
df
```

```python
df['To'] = df['To'].str.capitalize()
df
```

### Answer 4

```python
df = df.drop(['From_To'], axis=1)
df
```

```python
cols = list(df.columns)
```

```python
df = df[cols[3:]+cols[0:3]]
df
```

### Answer 5

```python
df[['delay_1','delay_2','delay_3']] = df['RecentDelays'].apply(pd.Series)
```

```python
df
```

```python
df=df.drop(['RecentDelays'],axis=1)
```

```python
df
```

```python
col = list(df.columns)
```

```python
df = df[col[0:3]+col[4:]+[col[3]]]
df
```

### Extra

```python
df['Airline'] = df['Airline'].str.strip('(!) <> (. 12. "   "')
```
